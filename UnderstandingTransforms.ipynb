```python
def imshow(img):
    plt.imshow(img)
    plt.show()
    
def hist(img):
    plt.hist(img)
    plt.show()
```

```python
import PIL.Image

image_path = osp.expanduser("~/data/datasets/CityScapes/CityScapes/leftImg8bit/train/aachen/aachen_000001_000019_leftImg8bit.png")
label_path = osp.expanduser("~/data/datasets/CityScapes/CityScapes/gtFine/train/aachen/aachen_000001_000019_gtFine_labelIds.png")
img = PIL.Image.open(image_path)
imshow(img)

lbl = PIL.Image.open(label_path)
imshow(lbl)
```

```python
from torchfcn.datasets.transforms import apply_transform, ImageTransformType, TransformDict, FlipType

max_transforms = len(TransformDict)

for i in range(max_transforms):
    transform_type = ImageTransformType(i+1)
    print(transform_type)
    new_image, new_lbl = apply_transform(img, lbl, transform_type, FlipType.Unknown, 256, 512)
    imshow(new_image)
    imshow(new_lbl)
```
